```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from sklearn import decomposition
from sklearn.decomposition import FastICA
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("Autism-Child-Data.csv")
data
```

```python
data.replace('?', np.nan, inplace = True)
data
```

```python
data.dropna(inplace = True)
data.drop("'4-11 years'", axis = 1, inplace = True)
```

```python
data = data.replace({'m': {"m" : 0, "f" : 1}, "no" : {"yes" : 1, "no" : 0}, "no.1" : {"yes" : 1, "no" : 0}, "no.2" : {"yes" : 1, "no" : 0}, "NO" : {"YES" : 1, "NO" : 0}})
```

```python
data = pd.get_dummies(data)
data
```

```python
scaler = StandardScaler()
data[['5']] = scaler.fit_transform(data[['5']])
data
```

```python
last_row = data["NO"]
```

```python
data.drop('NO', axis = 1, inplace = True)
```

```python
def model_RFC(feature, target):
    train_X, val_X, train_Y, val_Y = train_test_split(feature, target, test_size = 0.3, random_state = 42)
    model = RandomForestClassifier(n_estimators=500, random_state = 42, n_jobs = -1, max_depth = 10)
    model.fit(train_X, train_Y.values.ravel())
    preds = model.predict(val_X)
    score = accuracy_score(val_Y, preds)
    return score
```

```python
def model_DTC(feature, target):
    train_X, val_X, train_Y, val_Y = train_test_split(feature, target, test_size = 0.3, random_state = 42)
    model = DecisionTreeClassifier(max_depth = 20, random_state = 42)
    model.fit(train_X, train_Y)
    preds = model.predict(val_X)
    score = accuracy_score(val_Y, preds)
    return score
```

```python
def model_KNN(feature, target):
    train_X, val_X, train_Y, val_Y = train_test_split(feature, target, test_size = 0.3, random_state = 42)
    model = KNeighborsClassifier(n_neighbors = 100)
    model.fit(train_X, train_Y.values.ravel())
    preds = model.predict(val_X)
    score = accuracy_score(val_Y, preds)
    return score
```

```python
print("Decision Tree model accuracy without any feature extraction: ", str(model_DTC(data, last_row)))
print("Random Forest model accuracy without any feature extraction: ", str(model_RFC(data, last_row)))
print("KNN model accuracy without any feature extraction: ", str(model_KNN(data, last_row)))
```

```python
pca = decomposition.PCA(n_components=30, random_state=100)
pca.fit(data)
data_pca = pca.transform(data)
data_pca = pd.DataFrame(data_pca)
```

```python
print("Decision Tree model with PCA: ", str(model_DTC(data_pca, last_row)))
print("Random Forest model with PCA: ", str(model_RFC(data_pca, last_row)))
print("KNN model with PCA: ", str(model_KNN(data_pca, last_row)))
```

```python
ica = FastICA(n_components=42)
data_ica = ica.fit_transform(data)
data_ica = pd.DataFrame(data_ica)
```

```python
print("Decision Tree model with ICA: ", str(model_DTC(data_ica, last_row)))
print("Random Forest model with ICA: ", str(model_RFC(data_ica, last_row)))
print("KNN model with ICA: ", str(model_KNN(data_ica, last_row)))
```

```python
from sklearn.feature_selection import SelectPercentile as SP
selector = SP(percentile=50)
selector.fit(data,last_row.values.ravel())
data_mi = selector.transform(data)
data_mi = pd.DataFrame(data_mi)
```

```python
print("Decision Tree model with Mutual Information: ", str(model_DTC(data_mi, last_row)))
print("Random Forest model with Mutual Information: ", str(model_RFC(data_mi, last_row)))
print("KNN model with Mutual Information: ", str(model_KNN(data_mi, last_row)))
```
